```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Download necessary NLTK data
import nltk
nltk.download('stopwords', quiet=True)

def case_folding(text):
    """Converts text to lowercase.

    Args:
        text: The input string.

    Returns:
        The lowercase string.
    """
    if not isinstance(text, str):
        raise TypeError("Input must be a string.")
    return text.lower()

def clean_text(text):
    """Removes special characters, numbers, and punctuation from text.

    Args:
        text: The input string.

    Returns:
        The cleaned string.
    """
    if not isinstance(text, str):
        raise TypeError("Input must be a string.")
    text = re.sub(r'\d+', '', text) # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation)) # Remove punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove special characters
    return text.strip()

def tokenize_text(text):
    """Splits text into individual words (tokens).

    Args:
        text: The input string.

    Returns:
        A list of tokens.
    """
    if not isinstance(text, str):
        raise TypeError("Input must be a string.")
    return text.split()

def remove_stopwords(tokens, language='english'):
    """Removes common words (stopwords) from a list of tokens.

    Args:
        tokens: A list of tokens.
        language: The language of the stopwords ('english' or 'indonesian').

    Returns:
        A list of tokens with stopwords removed.
    """
    if not isinstance(tokens, list):
        raise TypeError("Input must be a list of tokens.")
    if language.lower() == 'english':
        stop_words = set(stopwords.words('english'))
    elif language.lower() == 'indonesian':
        stop_words = set(stopwords.words('indonesian'))
    else:
        raise ValueError("Unsupported language. Choose 'english' or 'indonesian'.")

    return [token for token in tokens if token not in stop_words]

def stem_tokens(tokens, language='english'):
    """Reduces words to their root form.

    Args:
        tokens: A list of tokens.
        language: The language for stemming ('english' or 'indonesian').

    Returns:
        A list of stemmed tokens.
    """
    if not isinstance(tokens, list):
        raise TypeError("Input must be a list of tokens.")

    if language.lower() == 'english':
        stemmer = PorterStemmer()
        return [stemmer.stem(token) for token in tokens]
    elif language.lower() == 'indonesian':
        factory = StemmerFactory()
        stemmer = factory.create_stemmer()
        return [stemmer.stem(token) for token in tokens]
    else:
        raise ValueError("Unsupported language. Choose 'english' or 'indonesian'.")

def normalize_tokens(tokens, language='english'):
    """Normalizes informal word forms.

    Args:
        tokens: A list of tokens.
        language: The language for normalization ('english' or 'indonesian').

    Returns:
        A list of normalized tokens.
    """
    if not isinstance(tokens, list):
        raise TypeError("Input must be a list of tokens.")

    # Example informal word dictionaries (can be expanded)
    english_informal_dict = {
        'gonna': 'going to',
        'wanna': 'want to',
        'btw': 'by the way'
    }
    indonesian_informal_dict = {
        'bgt': 'banget',
        'gak': 'tidak',
        'udah': 'sudah'
    }

    if language.lower() == 'english':
        normalization_dict = english_informal_dict
    elif language.lower() == 'indonesian':
        normalization_dict = indonesian_informal_dict
    else:
        raise ValueError("Unsupported language. Choose 'english' or 'indonesian'.")

    return [normalization_dict.get(token, token) for token in tokens]
```

## Apply preprocessing

### Subtask:
Apply the defined preprocessing functions to the `full_text` column of both `df_education` and `df_pendidikan` DataFrames.


**Reasoning**:
Define a function `preprocess_text` that takes a text string and a language as input, and applies the preprocessing steps in order. Then apply this function to the relevant columns of the dataframes for both languages.



```python
def preprocess_text(text, language='english'):
    """
    Applies a sequence of text preprocessing steps.

    Args:
        text: The input string.
        language: The language of the text ('english' or 'indonesian').

    Returns:
        A list of preprocessed tokens.
    """
    try:
        # Case Folding
        text = case_folding(text)
        # Cleaning
        text = clean_text(text)
        # Tokenization
        tokens = tokenize_text(text)
        # Stopword Removal
        tokens = remove_stopwords(tokens, language=language)
        # Stemming
        tokens = stem_tokens(tokens, language=language)
        # Normalization
        tokens = normalize_tokens(tokens, language=language)
        return tokens
    except Exception as e:
        print(f"Error processing text: {text[:50]}...")
        print(f"Error details: {e}")
        return []

# Apply preprocessing to df_education (English)
df_education['processed_text'] = df_education['full_text'].apply(lambda x: preprocess_text(x, language='english'))

# Apply preprocessing to df_pendidikan (Indonesian)
df_pendidikan['processed_text'] = df_pendidikan['full_text'].apply(lambda x: preprocess_text(x, language='indonesian'))

# Display the head of both dataframes
print("Processed df_education:")
display(df_education.head())

print("\nProcessed df_pendidikan:")
display(df_pendidikan.head())
```

## Visualize results

### Subtask:
Generate visualizations such as word clouds and word frequency plots for the preprocessed text of both DataFrames.


**Reasoning**:
Import necessary visualization libraries.



```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
```

**Reasoning**:
Generate word clouds and word frequency plots for the processed text of both DataFrames.



```python
# Function to generate and display word cloud and word frequency plot
def visualize_processed_text(df, title, language='english'):
    """
    Generates and displays a word cloud and word frequency plot for processed text.

    Args:
        df: DataFrame containing 'processed_text' column.
        title: Title for the visualizations.
        language: Language of the text ('english' or 'indonesian').
    """
    # Join tokens into a single string
    all_tokens = [token for sublist in df['processed_text'] for token in sublist]
    text = " ".join(all_tokens)

    # Generate Word Cloud
    print(f"Generating Word Cloud for {title}...")
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text)

    plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f'Word Cloud for {title}')
    plt.show()

    # Create Frequency Distribution
    print(f"Generating Word Frequency Plot for {title}...")
    fdist = Counter(all_tokens)
    top_words = fdist.most_common(20) # Get top 20 most common words

    # Plot Word Frequency
    plt.figure(figsize=(12, 6))
    sns.barplot(x=[word[0] for word in top_words], y=[word[1] for word in top_words], palette='viridis')
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Top 20 Most Frequent Words in {title}')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Visualize for df_education
visualize_processed_text(df_education, 'Tweet Education (English)', language='english')

# Visualize for df_pendidikan
visualize_processed_text(df_pendidikan, 'Tweet Pendidikan (Indonesian)', language='indonesian')
```
